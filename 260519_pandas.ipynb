### 타이타닉 기초 탐색

```python
import pandas as pd
TITANIC_URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(TITANIC_URL)
```

```python
# 1. 행 수, 열 수 출력
df.shape
```

```python
# 2. 컬럼별 타입·결측치 확인
df.info()
```

```python
# 3. 생존자(survived=1)만 필터링
df[df["Survived"] == 1]
```

```python
# 4. 성별에 따른 생존율 계산
df.groupby("Sex")[["Survived"]].mean()
```

### 그룹별 분석

```python
# 6. 등급별 평균 나이·요금·생존율 (agg 사용)
df.groupby("Pclass")[["Age","Fare","Survived"]].agg(생존율=('Survived','mean'),
                                                    평균나이=('Age','mean'),
                                                    요금=('Fare','mean'))
```

```python
# 7. 1등급 여성 승객 이름 목록
first_f = df[(df['Pclass'] == 1) & (df['Sex'] == 'female')]
first_f[['Name']]
```

```python
# 8. 35세 초과 + 1등급 + 남성 승객 수
df[(df['Pclass'] == 1) & (df['Age'] > 35)].groupby('Sex').agg(승객수=('Sex','count')).loc[['male']]
```

```python
# 9. 탑승항(embark_town)별 생존율 비교
df.groupby('Embarked')['Survived'].value_counts(normalize=True)
```

# 쇼핑몰 데이터

```python
import numpy as np
np.random.seed(42)
n = 150
shop = pd.DataFrame({
    'customer': np.random.choice(['김민준','이서현','박지우','최하은','정도윤'], n),
    'category': np.random.choice(['의류','전자기기','식품','생활용품','뷰티'], n),
    'quantity': np.random.randint(1, 6, n),
    'price':    np.random.choice([15000,29000,45000,89000,120000], n),
    'rating':   np.random.choice([1,2,3,4,5], n, p=[0.05,0.10,0.20,0.35,0.30]),
    'returned': np.random.choice([True, False], n, p=[0.1, 0.9])
})
shop.loc[np.random.choice(shop.index, 8, replace=False), 'rating'] = float('nan')
shop.to_csv('shop_orders.csv', index=False)
```

```python
df = pd.read_csv('shop_orders.csv')
df
```

```python
# Q1. 카테고리별 총 매출 내림차순
df['sales'] = df['price']*df['quantity']
df.groupby('category')[['sales']].sum().sort_values(by='sales',ascending=False)
```

```python
df.groupby('Embarked')['Survived'].value_counts(normalize=True)
```

```python
# Q2. 반품률 가장 높은 카테고리
df.groupby('category')[['returned']].sum()[['returned']].idxmax()
```

```python
# Q3. rating 결측치를 category별 평균으로 채우기
df_copy = df.copy()
df_copy.isnull().sum()
df_copy['rating'].fillna(df_copy.groupby('category')['rating'].mean())
```

```python
# Q4. 고객별 총 주문 금액 TOP 3
df.groupby('customer')[['sales']].sum().nlargest(3,'sales')
```
