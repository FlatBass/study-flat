### DataFrame 만들기

```python
import pandas as pd

students = pd.DataFrame({
    'name':    ['김민준', '이서현', '박지우', '최하은', '정도윤'],
    'age':     [22, 23, 22, 24, 23],
    'math':    [85, 92, 78, 95, 88],
    'english': [76, 88, 91, 70, 85],
    'grade':   ['B', 'A', 'B', 'A', 'A']
})
```

### 기본 탐색

```python
# head(행개수) : default 5
students.head()
```

```python
# shape : (행, 열) 튜플로 반환
students.shape
```

```python
# dtype : 컬럼별 타입 확인
students.dtypes
```

```python
# info() : 타입 + 결측치 수
# RangeIndex: 행수, Data columns(컬럼수), dtypes: 컬럼 데이터 타입, memory usage : 용량
# Non-Null Count : 결측치 개수
students.info()
```

```python
# describe() : 숫자형 통계
students.describe()
```
